From scratch development of K-Nearest Neighbors (KNN)

```python
# ============================================================================
# K-NEAREST NEIGHBORS (KNN) FROM SCRATCH - GOOGLE COLAB VERSION
# ============================================================================

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix, ConfusionMatrixDisplay
from sklearn.preprocessing import StandardScaler
from sklearn.neighbors import KNeighborsClassifier  # Pour comparaison
import time
import os
import warnings
warnings.filterwarnings('ignore')

print("="*80)
print("K-NEAREST NEIGHBORS (KNN) FROM SCRATCH - GOOGLE COLAB VERSION")
print("="*80)

# ============================================================================
# 1. CHARGEMENT DES DONN√âES
# ============================================================================

print("\nüìÅ 1. CHARGEMENT DES DONN√âES")

# Option 1: Essayer de monter Google Drive
try:
    from google.colab import drive
    print("Tentative de montage de Google Drive...")
    drive.mount('/content/drive', force_remount=False)

    # Chercher les fichiers
    data_dir = "/content/drive/MyDrive/"

    # Essayer diff√©rents chemins
    possible_paths = [
        "/content/drive/MyDrive/processed_balanced/",

    ]

    files_found = False
    for path in possible_paths:
        try:
            print(f"Tentative: {path}")
            X_train = pd.read_csv(f"{path}X_train_scaled.csv")
            y_train = pd.read_csv(f"{path}y_train.csv")['fire']
            X_test = pd.read_csv(f"{path}X_test_scaled.csv")
            y_test = pd.read_csv(f"{path}y_test.csv")['fire']
            files_found = True
            print(f"‚úÖ Dataset charg√© depuis: {path}")
            break
        except Exception as e:
            print(f"‚ùå Erreur: {e}")
            continue

    if not files_found:
        raise FileNotFoundError("Fichiers non trouv√©s sur Google Drive")

except Exception as e:
    print(f"‚ùå Google Drive non disponible ou fichiers non trouv√©s: {e}")
    print("Cr√©ation d'un dataset synth√©tique pour d√©monstration...")

    # Option 2: Cr√©er un dataset synth√©tique
    n_samples = 2000  # R√©duit pour KNN (plus rapide)
    n_features = 10

    # G√©n√©rer deux clusters distincts pour les deux classes
    np.random.seed(42)

    # Classe 0: centre √† (0, 0, ...)
    n_class0 = n_samples // 2
    X_class0 = np.random.randn(n_class0, n_features) * 1.0

    # Classe 1: centre √† (2, 2, ...)
    n_class1 = n_samples - n_class0
    X_class1 = np.random.randn(n_class1, n_features) * 1.0 + 2.0

    # Combiner
    X = np.vstack([X_class0, X_class1])
    y = np.hstack([np.zeros(n_class0), np.ones(n_class1)])

    # M√©langer
    indices = np.random.permutation(n_samples)
    X = X[indices]
    y = y[indices]

    # Split train/test
    split_idx = int(0.8 * n_samples)
    X_train = pd.DataFrame(X[:split_idx], columns=[f'feature_{i}' for i in range(n_features)])
    X_test = pd.DataFrame(X[split_idx:], columns=[f'feature_{i}' for i in range(n_features)])
    y_train = pd.Series(y[:split_idx], name='fire')
    y_test = pd.Series(y[split_idx:], name='fire')

    print(f"‚úÖ Dataset synth√©tique cr√©√© avec succ√®s!")
    print(f"   Shape: {X_train.shape}, 2 clusters bien s√©par√©s")

# Afficher les informations sur les donn√©es
print(f"\nüìä DIMENSIONS :")
print(f"   X_train : {X_train.shape}")
print(f"   y_train : {y_train.shape}")
print(f"   X_test  : {X_test.shape}")
print(f"   y_test  : {y_test.shape}")

# V√©rifier les valeurs NaN
print(f"\nüîç V√âRIFICATION DES DONN√âES :")
print(f"   NaN dans X_train: {X_train.isna().sum().sum()}")
print(f"   NaN dans X_test: {X_test.isna().sum().sum()}")

# Remplacer les NaN si n√©cessaire
if X_train.isna().sum().sum() > 0:
    print("‚ö†Ô∏è  Remplissage des valeurs NaN par la moyenne...")
    X_train = X_train.fillna(X_train.mean())
    X_test = X_test.fillna(X_test.mean())

print(f"\nüìà DISTRIBUTION DES CLASSES :")
train_counts = y_train.value_counts()
test_counts = y_test.value_counts()
print(f"   Training - Classe 0: {train_counts[0]:,} ({train_counts[0]/len(y_train)*100:.1f}%)")
print(f"   Training - Classe 1: {train_counts[1]:,} ({train_counts[1]/len(y_train)*100:.1f}%)")
print(f"   Test     - Classe 0: {test_counts[0]:,} ({test_counts[0]/len(y_test)*100:.1f}%)")
print(f"   Test     - Classe 1: {test_counts[1]:,} ({test_counts[1]/len(y_test)*100:.1f}%)")

# Convertir en numpy arrays
X_train_np = X_train.values.astype(np.float32)  # float32 pour plus de rapidit√©
y_train_np = y_train.values.astype(np.int32)
X_test_np = X_test.values.astype(np.float32)
y_test_np = y_test.values.astype(np.int32)

print(f"\n‚úÖ Donn√©es pr√™tes pour l'entra√Ænement!")

# ============================================================================
# 2. IMPL√âMENTATION OPTIMIS√âE DE KNN FROM SCRATCH
# ============================================================================

print("\nüîß 2. IMPL√âMENTATION OPTIMIS√âE DE KNN FROM SCRATCH")

class KNNFromScratch:
    """K-Nearest Neighbors impl√©ment√© from scratch avec optimisations"""

    def __init__(self, k=5, distance_metric='euclidean', weights='uniform'):
        """
        Parameters:
        -----------
        k : int, nombre de voisins √† consid√©rer
        distance_metric : str, m√©trique de distance ('euclidean', 'manhattan', 'minkowski')
        weights : str, poids des voisins ('uniform' ou 'distance')
        """
        self.k = k
        self.distance_metric = distance_metric
        self.weights = weights
        self.X_train = None
        self.y_train = None
        self.n_classes = None

    def _compute_distance(self, x1, x2):
        """Calcule la distance entre deux points"""
        if self.distance_metric == 'euclidean':
            return np.sqrt(np.sum((x1 - x2) ** 2))
        elif self.distance_metric == 'manhattan':
            return np.sum(np.abs(x1 - x2))
        elif self.distance_metric == 'minkowski':
            # Minkowski avec p=3
            return np.sum(np.abs(x1 - x2) ** 3) ** (1/3)
        else:
            raise ValueError(f"M√©trique de distance non support√©e: {self.distance_metric}")

    def _compute_distances_vectorized(self, X):
        """Calcule les distances de mani√®re vectoris√©e (plus rapide)"""
        if self.distance_metric == 'euclidean':
            # Utiliser l'astuce: ||a - b||¬≤ = ||a||¬≤ + ||b||¬≤ - 2a¬∑b
            X_norm = np.sum(X ** 2, axis=1).reshape(1, -1)
            train_norm = np.sum(self.X_train ** 2, axis=1).reshape(-1, 1)
            distances = np.sqrt(X_norm + train_norm - 2 * np.dot(self.X_train, X.T))
            return distances.T
        else:
            # Pour les autres m√©triques, calculer de mani√®re plus simple
            n_test = X.shape[0]
            n_train = self.X_train.shape[0]
            distances = np.zeros((n_test, n_train))

            for i in range(n_test):
                for j in range(n_train):
                    distances[i, j] = self._compute_distance(X[i], self.X_train[j])

            return distances

    def fit(self, X, y):
        """Stocke les donn√©es d'entra√Ænement (algorithme paresseux)"""
        print(f"‚è≥ Stockage de {X.shape[0]:,} observations d'entra√Ænement...")
        self.X_train = X.copy()
        self.y_train = y.copy()
        self.n_classes = len(np.unique(y))
        print(f"‚úÖ Donn√©es stock√©es - KNN pr√™t pour les pr√©dictions")
        return self

    def predict_single(self, x):
        """Pr√©dit la classe pour un seul √©chantillon"""
        # Calculer les distances √† tous les points d'entra√Ænement
        distances = []
        for i in range(len(self.X_train)):
            dist = self._compute_distance(x, self.X_train[i])
            distances.append((dist, self.y_train[i]))

        # Trier par distance
        distances.sort(key=lambda x: x[0])

        # Prendre les k plus proches voisins
        k_neighbors = distances[:self.k]

        # Calculer les votes
        if self.weights == 'uniform':
            votes = {}
            for _, label in k_neighbors:
                votes[label] = votes.get(label, 0) + 1
            return max(votes.items(), key=lambda x: x[1])[0]

        elif self.weights == 'distance':
            # Poids inversement proportionnels √† la distance
            weighted_votes = {}
            for dist, label in k_neighbors:
                # √âviter la division par z√©ro
                weight = 1.0 / (dist + 1e-10)
                weighted_votes[label] = weighted_votes.get(label, 0) + weight
            return max(weighted_votes.items(), key=lambda x: x[1])[0]

    def predict(self, X):
        """Pr√©dit les classes pour plusieurs √©chantillons"""
        print(f"‚è≥ Pr√©dictions KNN (k={self.k}) sur {X.shape[0]:,} observations...")
        start_time = time.time()

        predictions = np.zeros(len(X), dtype=self.y_train.dtype)

        # Calculer toutes les distances de mani√®re vectoris√©e si possible
        if self.distance_metric == 'euclidean' and len(X) > 1:
            distances = self._compute_distances_vectorized(X)

            for i in range(len(X)):
                # Obtenir les indices des k plus proches voisins
                k_indices = np.argsort(distances[i])[:self.k]
                k_labels = self.y_train[k_indices]

                if self.weights == 'uniform':
                    # Vote majoritaire simple
                    unique, counts = np.unique(k_labels, return_counts=True)
                    predictions[i] = unique[np.argmax(counts)]

                elif self.weights == 'distance':
                    # Vote pond√©r√© par l'inverse de la distance
                    k_distances = distances[i, k_indices]
                    weights = 1.0 / (k_distances + 1e-10)

                    weighted_votes = {}
                    for label, weight in zip(k_labels, weights):
                        weighted_votes[label] = weighted_votes.get(label, 0) + weight

                    predictions[i] = max(weighted_votes.items(), key=lambda x: x[1])[0]
        else:
            # Version non-vectoris√©e pour les petites datasets ou autres m√©triques
            for i in range(len(X)):
                predictions[i] = self.predict_single(X[i])

                # Afficher la progression
                if (i + 1) % max(1, len(X) // 10) == 0:
                    print(f"   Progress: {i+1}/{len(X)} ({((i+1)/len(X)*100):.1f}%)")

        prediction_time = time.time() - start_time
        print(f"‚úÖ Pr√©dictions termin√©es en {prediction_time:.2f} secondes")

        return predictions

    def predict_proba(self, X):
        """Retourne les probabilit√©s pour chaque classe"""
        print(f"‚è≥ Calcul des probabilit√©s KNN sur {X.shape[0]:,} observations...")

        probabilities = np.zeros((len(X), self.n_classes))

        # Calculer toutes les distances de mani√®re vectoris√©e si possible
        if self.distance_metric == 'euclidean' and len(X) > 1:
            distances = self._compute_distances_vectorized(X)

            for i in range(len(X)):
                # Obtenir les indices des k plus proches voisins
                k_indices = np.argsort(distances[i])[:self.k]
                k_labels = self.y_train[k_indices]

                if self.weights == 'uniform':
                    # Probabilit√©s bas√©es sur les fr√©quences
                    for label in range(self.n_classes):
                        probabilities[i, label] = np.sum(k_labels == label) / self.k

                elif self.weights == 'distance':
                    # Probabilit√©s pond√©r√©es par l'inverse de la distance
                    k_distances = distances[i, k_indices]
                    weights = 1.0 / (k_distances + 1e-10)

                    total_weight = np.sum(weights)
                    for label in range(self.n_classes):
                        label_mask = k_labels == label
                        if np.any(label_mask):
                            probabilities[i, label] = np.sum(weights[label_mask]) / total_weight
                        else:
                            probabilities[i, label] = 0.0

        return probabilities

    def get_params(self):
        """Retourne les param√®tres du mod√®le"""
        return {
            'k': self.k,
            'distance_metric': self.distance_metric,
            'weights': self.weights,
            'n_train_samples': len(self.X_train) if self.X_train is not None else 0,
            'n_classes': self.n_classes
        }

# ============================================================================
# 3. ENTRA√éNEMENT ET √âVALUATION DE KNN
# ============================================================================

print("\n" + "="*80)
print("üå± 3. ENTRA√éNEMENT ET √âVALUATION DE KNN")
print("="*80)

# Cr√©er et entra√Æner le mod√®le KNN
print(f"\nüîß Cr√©ation du mod√®le KNN...")
knn_scratch = KNNFromScratch(
    k=5,                    # Nombre de voisins
    distance_metric='euclidean',  # M√©trique de distance
    weights='uniform'       # Poids uniformes
)

# Entra√Æner (juste stocker les donn√©es)
knn_scratch.fit(X_train_np, y_train_np)

# Afficher les param√®tres
params = knn_scratch.get_params()
print(f"\nüìã PARAM√àTRES DU MOD√àLE :")
print(f"   ‚Ä¢ k (nombre de voisins) : {params['k']}")
print(f"   ‚Ä¢ Distance metric : {params['distance_metric']}")
print(f"   ‚Ä¢ Weights : {params['weights']}")
print(f"   ‚Ä¢ Training samples : {params['n_train_samples']:,}")
print(f"   ‚Ä¢ Number of classes : {params['n_classes']}")

# Pr√©dictions
print(f"\nüéØ PR√âDICTIONS SUR LE JEU DE TEST...")
y_pred_knn = knn_scratch.predict(X_test_np)

# Calcul des m√©triques
accuracy_knn = accuracy_score(y_test_np, y_pred_knn)

print(f"\nüèÜ R√âSULTATS KNN FROM SCRATCH :")
print("="*60)
print(f"   Accuracy : {accuracy_knn:.4f}")
print(f"   k value  : {knn_scratch.k}")
print(f"   Test samples : {len(y_test_np):,}")

# Rapport de classification
print(f"\nüìã RAPPORT DE CLASSIFICATION :")
print("="*60)
print(classification_report(y_test_np, y_pred_knn,
                           target_names=['Pas de feu (0)', 'Feu (1)']))

# Matrice de confusion
print(f"\nüéØ MATRICE DE CONFUSION :")
print("="*60)
cm_knn = confusion_matrix(y_test_np, y_pred_knn)
print(f"\n{cm_knn}")
print(f"\n‚Ä¢ Vrais n√©gatifs (TN): {cm_knn[0, 0]}")
print(f"‚Ä¢ Faux positifs (FP): {cm_knn[0, 1]}")
print(f"‚Ä¢ Faux n√©gatifs (FN): {cm_knn[1, 0]}")
print(f"‚Ä¢ Vrais positifs (TP): {cm_knn[1, 1]}")

# Calculer les m√©triques additionnelles
tn, fp, fn, tp = cm_knn.ravel()
accuracy_calculated = (tp + tn) / (tp + tn + fp + fn)
precision = tp / (tp + fp) if (tp + fp) > 0 else 0
recall = tp / (tp + fn) if (tp + fn) > 0 else 0
f1_score = 2 * (precision * recall) / (precision + recall) if (precision + recall) > 0 else 0

print(f"\nüìà M√âTRIQUES D√âTAILL√âES :")
print(f"   ‚Ä¢ Accuracy (calcul√©e) : {accuracy_calculated:.4f}")
print(f"   ‚Ä¢ Precision : {precision:.4f}")
print(f"   ‚Ä¢ Recall (Sensibilit√©) : {recall:.4f}")
print(f"   ‚Ä¢ F1-Score : {f1_score:.4f}")
print(f"   ‚Ä¢ Sp√©cificit√© : {tn/(tn+fp):.4f}" if (tn+fp) > 0 else "   ‚Ä¢ Sp√©cificit√© : N/A")

# Visualisation de la matrice de confusion
fig, ax = plt.subplots(figsize=(8, 6))
disp = ConfusionMatrixDisplay(confusion_matrix=cm_knn, display_labels=['Pas de feu', 'Feu'])
disp.plot(ax=ax, cmap='Blues', values_format='d')
plt.title(f'Matrice de Confusion - KNN from Scratch (k={knn_scratch.k})', fontsize=14, pad=15)
plt.tight_layout()
plt.show()

# ============================================================================
# 4. COMPARAISON AVEC SCIKIT-LEARN KNN
# ============================================================================

print("\n" + "="*80)
print("üîç 4. COMPARAISON AVEC SCIKIT-LEARN KNN")
print("="*80)

print(f"\nüîß Entra√Ænement de scikit-learn KNN...")

# Utiliser les m√™mes param√®tres
knn_sklearn = KNeighborsClassifier(
    n_neighbors=knn_scratch.k,
    metric=knn_scratch.distance_metric,
    weights=knn_scratch.weights
)

start_time = time.time()
knn_sklearn.fit(X_train_np, y_train_np)
sklearn_train_time = time.time() - start_time

start_time = time.time()
y_pred_sklearn = knn_sklearn.predict(X_test_np)
sklearn_pred_time = time.time() - start_time

accuracy_sklearn = accuracy_score(y_test_np, y_pred_sklearn)

print(f"\nüìä COMPARAISON :")
print("="*60)
print(f"{'M√©trique':<25} {'From Scratch':<15} {'Scikit-learn':<15}")
print("-"*60)
print(f"{'Accuracy':<25} {accuracy_knn:.4f}          {accuracy_sklearn:.4f}")
print(f"{'Training time':<25} {'0.00s (lazy)':<15} {sklearn_train_time:.2f}s")
print(f"{'Prediction time':<25} {'N/A':<15} {sklearn_pred_time:.2f}s")
print(f"{'k (voisins)':<25} {knn_scratch.k:<15} {knn_sklearn.n_neighbors}")
print(f"{'Distance metric':<25} {knn_scratch.distance_metric:<15} {knn_sklearn.metric}")

# Diff√©rence de performance
diff = abs(accuracy_knn - accuracy_sklearn)
print(f"\nüìà DIFF√âRENCE D'ACCURACY : {diff:.4f}")
if diff < 0.01:
    print("‚úÖ Les deux impl√©mentations donnent des r√©sultats similaires")
else:
    print("‚ö†Ô∏è  Diff√©rence notable entre les impl√©mentations")

# Visualisation comparative
fig, ax = plt.subplots(figsize=(10, 6))
models = ['KNN From Scratch', 'KNN Scikit-learn']
accuracies = [accuracy_knn, accuracy_sklearn]
colors = ['steelblue', 'darkorange']

bars = ax.bar(models, accuracies, color=colors, alpha=0.8)
ax.set_ylabel('Accuracy', fontsize=12)
ax.set_title(f'Comparaison des performances KNN (k={knn_scratch.k})', fontsize=14, pad=15)
ax.grid(True, alpha=0.3, axis='y')
ax.set_ylim([0, 1.0])

# Ajouter les valeurs sur les barres
for bar, acc in zip(bars, accuracies):
    height = bar.get_height()
    ax.text(bar.get_x() + bar.get_width()/2., height + 0.01,
            f'{acc:.4f}', ha='center', va='bottom', fontsize=11)

plt.tight_layout()
plt.show()

# ============================================================================
# 5. ANALYSE DE L'IMPACT DU PARAM√àTRE K
# ============================================================================

print("\n" + "="*80)
print("üìä 5. ANALYSE DE L'IMPACT DU PARAM√àTRE K")
print("="*80)

# Tester diff√©rentes valeurs de k
print(f"\nüîß Test avec diff√©rentes valeurs de k...")

k_values = [1, 3, 5, 7, 9, 11, 15, 21]
results_k = []

for k in k_values:
    print(f"\n‚è≥ Test avec k = {k}...")

    # Mod√®le from scratch
    knn_test = KNNFromScratch(
        k=k,
        distance_metric='euclidean',
        weights='uniform'
    )

    knn_test.fit(X_train_np, y_train_np)
    y_pred_test = knn_test.predict(X_test_np)
    accuracy_test = accuracy_score(y_test_np, y_pred_test)

    # Mod√®le scikit-learn pour comparaison
    knn_sk_test = KNeighborsClassifier(n_neighbors=k, metric='euclidean', weights='uniform')
    knn_sk_test.fit(X_train_np, y_train_np)
    y_pred_sk_test = knn_sk_test.predict(X_test_np)
    accuracy_sk_test = accuracy_score(y_test_np, y_pred_sk_test)

    results_k.append({
        'k': k,
        'accuracy_scratch': accuracy_test,
        'accuracy_sklearn': accuracy_sk_test,
        'difference': abs(accuracy_test - accuracy_sk_test)
    })

    print(f"   From Scratch: {accuracy_test:.4f}, Scikit-learn: {accuracy_sk_test:.4f}, Diff: {abs(accuracy_test - accuracy_sk_test):.4f}")

# Graphique
results_k_df = pd.DataFrame(results_k)
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(14, 5))

# Graphique 1: Accuracy vs k
ax1.plot(results_k_df['k'], results_k_df['accuracy_scratch'], 'o-', linewidth=2, markersize=8,
         color='steelblue', label='From Scratch')
ax1.plot(results_k_df['k'], results_k_df['accuracy_sklearn'], 's--', linewidth=2, markersize=8,
         color='darkorange', label='Scikit-learn')
ax1.set_xlabel('Valeur de k (nombre de voisins)', fontsize=11)
ax1.set_ylabel('Accuracy', fontsize=11)
ax1.set_title('Accuracy vs k (KNN)', fontsize=13, pad=15)
ax1.grid(True, alpha=0.3)
ax1.set_xticks(k_values)
ax1.legend()

# Graphique 2: Diff√©rence vs k
ax2.bar(results_k_df['k'], results_k_df['difference'], color='coral', alpha=0.7)
ax2.set_xlabel('Valeur de k (nombre de voisins)', fontsize=11)
ax2.set_ylabel('Diff√©rence d\'accuracy', fontsize=11)
ax2.set_title('Diff√©rence entre impl√©mentations vs k', fontsize=13, pad=15)
ax2.grid(True, alpha=0.3, axis='y')
ax2.set_xticks(k_values)

plt.tight_layout()
plt.show()

# Trouver le meilleur k
best_k_idx = results_k_df['accuracy_scratch'].idxmax()
best_k = results_k_df.loc[best_k_idx, 'k']
best_accuracy = results_k_df.loc[best_k_idx, 'accuracy_scratch']

print(f"\nüéØ MEILLEUR PARAM√àTRE K :")
print(f"   ‚Ä¢ k optimal : {best_k}")
print(f"   ‚Ä¢ Accuracy optimale : {best_accuracy:.4f}")
print(f"   ‚Ä¢ Am√©lioration vs k=5 : {(best_accuracy - accuracy_knn):.4f}")

# ============================================================================
# 6. ANALYSE DES DIFF√âRENTES M√âTRIQUES DE DISTANCE
# ============================================================================

print("\n" + "="*80)
print("üìè 6. ANALYSE DES DIFF√âRENTES M√âTRIQUES DE DISTANCE")
print("="*80)

# Tester diff√©rentes m√©triques de distance
print(f"\nüîß Test avec diff√©rentes m√©triques de distance (k=5)...")

distance_metrics = ['euclidean', 'manhattan']
results_metrics = []

for metric in distance_metrics:
    print(f"\n‚è≥ Test avec m√©trique = {metric}...")

    # Mod√®le from scratch
    knn_metric = KNNFromScratch(
        k=5,
        distance_metric=metric,
        weights='uniform'
    )

    knn_metric.fit(X_train_np, y_train_np)
    y_pred_metric = knn_metric.predict(X_test_np)
    accuracy_metric = accuracy_score(y_test_np, y_pred_metric)

    results_metrics.append({
        'metric': metric,
        'accuracy': accuracy_metric
    })

    print(f"   Accuracy avec {metric}: {accuracy_metric:.4f}")

# Graphique
results_metrics_df = pd.DataFrame(results_metrics)
plt.figure(figsize=(10, 6))
bars = plt.bar(results_metrics_df['metric'], results_metrics_df['accuracy'],
               color=['steelblue', 'lightcoral'], alpha=0.8)
plt.xlabel('M√©trique de distance', fontsize=12)
plt.ylabel('Accuracy', fontsize=12)
plt.title('Impact de la m√©trique de distance sur les performances KNN (k=5)', fontsize=14, pad=15)
plt.grid(True, alpha=0.3, axis='y')
plt.ylim([0, 1.0])

# Ajouter les valeurs sur les barres
for bar, acc in zip(bars, results_metrics_df['accuracy']):
    height = bar.get_height()
    plt.text(bar.get_x() + bar.get_width()/2., height + 0.01,
             f'{acc:.4f}', ha='center', va='bottom', fontsize=11)

plt.tight_layout()
plt.show()

# ============================================================================
# 7. ANALYSE DES DIFF√âRENTS TYPES DE POIDS
# ============================================================================

print("\n" + "="*80)
print("‚öñÔ∏è 7. ANALYSE DES DIFF√âRENTS TYPES DE POIDS")
print("="*80)

# Tester diff√©rents types de poids
print(f"\nüîß Test avec diff√©rents types de poids (k=5, distance=euclidean)...")

weights_types = ['uniform', 'distance']
results_weights = []

for weight in weights_types:
    print(f"\n‚è≥ Test avec poids = {weight}...")

    # Mod√®le from scratch
    knn_weight = KNNFromScratch(
        k=5,
        distance_metric='euclidean',
        weights=weight
    )

    knn_weight.fit(X_train_np, y_train_np)
    y_pred_weight = knn_weight.predict(X_test_np)
    accuracy_weight = accuracy_score(y_test_np, y_pred_weight)

    results_weights.append({
        'weights': weight,
        'accuracy': accuracy_weight
    })

    print(f"   Accuracy avec poids {weight}: {accuracy_weight:.4f}")

# Graphique
results_weights_df = pd.DataFrame(results_weights)
plt.figure(figsize=(10, 6))
bars = plt.bar(results_weights_df['weights'], results_weights_df['accuracy'],
               color=['steelblue', 'lightgreen'], alpha=0.8)
plt.xlabel('Type de poids', fontsize=12)
plt.ylabel('Accuracy', fontsize=12)
plt.title('Impact du type de poids sur les performances KNN (k=5)', fontsize=14, pad=15)
plt.grid(True, alpha=0.3, axis='y')
plt.ylim([0, 1.0])

# Ajouter les valeurs sur les barres
for bar, acc in zip(bars, results_weights_df['accuracy']):
    height = bar.get_height()
    plt.text(bar.get_x() + bar.get_width()/2., height + 0.01,
             f'{acc:.4f}', ha='center', va='bottom', fontsize=11)

plt.tight_layout()
plt.show()

# ============================================================================
# 8. VISUALISATION DES D√âCISIONS KNN (2D)
# ============================================================================

print("\n" + "="*80)
print("üéØ 8. VISUALISATION DES D√âCISIONS KNN (2D)")
print("="*80)

# Visualiser les fronti√®res de d√©cision (en 2D pour la visualisation)
if X_train_np.shape[1] >= 2:
    print(f"\nüîç Visualisation des fronti√®res de d√©cision (2 premi√®res features)...")

    # Prendre seulement les 2 premi√®res features pour la visualisation
    X_train_2d = X_train_np[:, :2]
    X_test_2d = X_test_np[:, :2]

    # Cr√©er un meshgrid pour la visualisation
    x_min, x_max = X_train_2d[:, 0].min() - 1, X_train_2d[:, 0].max() + 1
    y_min, y_max = X_train_2d[:, 1].min() - 1, X_train_2d[:, 1].max() + 1
    xx, yy = np.meshgrid(np.arange(x_min, x_max, 0.1),
                         np.arange(y_min, y_max, 0.1))

    # Cr√©er un mod√®le KNN simplifi√© pour la visualisation
    knn_viz = KNNFromScratch(k=5, distance_metric='euclidean', weights='uniform')
    knn_viz.fit(X_train_2d, y_train_np)

    # Pr√©dire sur le meshgrid
    print("‚è≥ Calcul des pr√©dictions pour la visualisation...")
    Z = knn_viz.predict(np.c_[xx.ravel(), yy.ravel()])
    Z = Z.reshape(xx.shape)

    # Cr√©er la visualisation
    plt.figure(figsize=(12, 8))

    # Tracer les contours
    plt.contourf(xx, yy, Z, alpha=0.4, cmap='RdBu_r')

    # Tracer les points d'entra√Ænement
    scatter_train = plt.scatter(X_train_2d[:, 0], X_train_2d[:, 1],
                               c=y_train_np, cmap='RdBu_r',
                               edgecolors='k', s=50, alpha=0.7,
                               label='Training data')

    # Tracer les points de test
    scatter_test = plt.scatter(X_test_2d[:, 0], X_test_2d[:, 1],
                              c=y_test_np, cmap='RdBu_r',
                              edgecolors='k', s=100, marker='s',
                              label='Test data')

    plt.xlabel(X_train.columns[0], fontsize=12)
    plt.ylabel(X_train.columns[1], fontsize=12)
    plt.title(f'Fronti√®res de d√©cision KNN (k=5) - Features 0 et 1', fontsize=14, pad=15)
    plt.legend()
    plt.colorbar(scatter_train, label='Classe (0=Pas de feu, 1=Feu)')
    plt.grid(True, alpha=0.3)
    plt.tight_layout()
    plt.show()

    print("‚úÖ Visualisation des fronti√®res de d√©cision termin√©e!")
else:
    print("‚ö†Ô∏è  Pas assez de features pour la visualisation 2D (minimum 2 features requises)")

# ============================================================================
# 9. CONCLUSION ET R√âCAPITULATIF
# ============================================================================

print("\n" + "="*80)
print("üéØ 9. CONCLUSION ET R√âCAPITULATIF")
print("="*80)

# R√©sum√© statistique
print(f"""
üìã R√âCAPITULATIF STATISTIQUE :
   ‚Ä¢ Dataset : {X_train_np.shape[0]:,} observations d'entra√Ænement
   ‚Ä¢ Features : {X_train_np.shape[1]}
   ‚Ä¢ Distribution des classes :
     - Classe 0 (Pas de feu) : {train_counts[0]} ({train_counts[0]/len(y_train)*100:.1f}%)
     - Classe 1 (Feu) : {train_counts[1]} ({train_counts[1]/len(y_train)*100:.1f}%)

üéØ PERFORMANCE DU MOD√àLE KNN :
   ‚Ä¢ KNN from scratch (k={knn_scratch.k}) : {accuracy_knn:.4f}
   ‚Ä¢ KNN scikit-learn (k={knn_scratch.k}) : {accuracy_sklearn:.4f}
   ‚Ä¢ Diff√©rence : {diff:.4f}
   ‚Ä¢ Meilleur k trouv√© : {best_k} (accuracy: {best_accuracy:.4f})

üìä IMPACT DES PARAM√àTRES :
   ‚Ä¢ M√©trique optimale : {results_metrics_df.loc[results_metrics_df['accuracy'].idxmax(), 'metric']}
   ‚Ä¢ Poids optimaux : {results_weights_df.loc[results_weights_df['accuracy'].idxmax(), 'weights']}
   ‚Ä¢ k optimal : {best_k}

üèÜ COMPARAISON DES IMPL√âMENTATIONS :
   ‚Ä¢ Votre impl√©mentation KNN est {'tr√®s proche' if diff < 0.01 else 'diff√©rente'} de scikit-learn
   ‚Ä¢ Algorithme : Lazy learning (pas d'entra√Ænement, seulement stockage)
   ‚Ä¢ Complexit√© en pr√©diction : O(n_train * n_test * n_features)

üöÄ RECOMMANDATIONS POUR VOTRE PROJET :
   ‚Ä¢ Choisir k = {best_k} pour les meilleures performances
   ‚Ä¢ Utiliser la m√©trique '{results_metrics_df.loc[results_metrics_df['accuracy'].idxmax(), 'metric']}'
   ‚Ä¢ Consid√©rer les poids '{results_weights_df.loc[results_weights_df['accuracy'].idxmax(), 'weights']}'
   ‚Ä¢ Pour les grands datasets, r√©duire les dimensions avec PCA
   ‚Ä¢ Normaliser les donn√©es pour de meilleures performances KNN

üìà POUR ALLER PLUS LOIN :
   ‚Ä¢ Impl√©menter KD-Tree ou Ball Tree pour acc√©l√©rer les recherches
   ‚Ä¢ Ajouter la validation crois√©e pour le choix optimal de k
   ‚Ä¢ Impl√©menter d'autres m√©triques de distance (cosine, mahalanobis)
   ‚Ä¢ Tester avec diff√©rentes strat√©gies de poids
""")

# Sauvegarde des r√©sultats dans un dictionnaire
knn_results_summary = {
    'dataset_info': {
        'n_train_samples': X_train_np.shape[0],
        'n_test_samples': X_test_np.shape[0],
        'n_features': X_train_np.shape[1],
        'class_distribution': {
            'train_class_0': int(train_counts[0]),
            'train_class_1': int(train_counts[1]),
            'test_class_0': int(test_counts[0]),
            'test_class_1': int(test_counts[1])
        }
    },
    'model_performance': {
        'accuracy_from_scratch': float(accuracy_knn),
        'accuracy_sklearn': float(accuracy_sklearn),
        'difference': float(diff),
        'best_k': int(best_k),
        'best_accuracy': float(best_accuracy)
    },
    'parameters_analysis': {
        'best_metric': results_metrics_df.loc[results_metrics_df['accuracy'].idxmax(), 'metric'],
        'best_weights': results_weights_df.loc[results_weights_df['accuracy'].idxmax(), 'weights']
    },
    'model_params': knn_scratch.get_params()
}

print(f"\nüìÅ R√©sum√© sauvegard√© dans le dictionnaire 'knn_results_summary'")
print(f"\n‚úÖ EX√âCUTION TERMIN√âE AVEC SUCC√àS!")
print(f"‚ú® Vous avez impl√©ment√© KNN from scratch avec {accuracy_knn:.1%} d'accuracy!")

# ============================================================================
# 10. BONUS : EXEMPLE DE PR√âDICTION D√âTAILL√âE
# ============================================================================

print("\n" + "="*80)
print("üîç 10. BONUS : EXEMPLE DE PR√âDICTION D√âTAILL√âE")
print("="*80)

# Prendre un √©chantillon de test
sample_idx = 0
sample = X_test_np[sample_idx]
true_label = y_test_np[sample_idx]
predicted_label = y_pred_knn[sample_idx]

print(f"\nüîç √âchantillon de test #{sample_idx}:")
print(f"   ‚Ä¢ Vraie classe : {true_label} ({'Feu' if true_label == 1 else 'Pas de feu'})")
print(f"   ‚Ä¢ Classe pr√©dite : {predicted_label} ({'Feu' if predicted_label == 1 else 'Pas de feu'})")
print(f"   ‚Ä¢ Pr√©diction correcte : {'‚úÖ Oui' if true_label == predicted_label else '‚ùå Non'}")

# Afficher les 5 plus proches voisins
print(f"\nüìä Les {knn_scratch.k} plus proches voisins:")

# Calculer les distances manuellement pour cet √©chantillon
distances = []
for i in range(len(knn_scratch.X_train)):
    dist = knn_scratch._compute_distance(sample, knn_scratch.X_train[i])
    distances.append((dist, knn_scratch.y_train[i], i))

# Trier et prendre les k plus proches
distances.sort(key=lambda x: x[0])
k_neighbors = distances[:knn_scratch.k]

for i, (dist, label, idx) in enumerate(k_neighbors):
    print(f"   {i+1}. Distance: {dist:.4f}, Classe: {label} ({'Feu' if label == 1 else 'Pas de feu'})")

# Calculer le vote (pour information)
if knn_scratch.weights == 'uniform':
    votes = {}
    for _, label, _ in k_neighbors:
        votes[label] = votes.get(label, 0) + 1
    print(f"\nüó≥Ô∏è  Votes : Classe 0 = {votes.get(0, 0)}, Classe 1 = {votes.get(1, 0)}")
    print(f"   ‚Üí D√©cision : Classe {predicted_label} (vote majoritaire)")

elif knn_scratch.weights == 'distance':
    weighted_votes = {}
    for dist, label, _ in k_neighbors:
        weight = 1.0 / (dist + 1e-10)
        weighted_votes[label] = weighted_votes.get(label, 0) + weight
    print(f"\n‚öñÔ∏è  Votes pond√©r√©s : Classe 0 = {weighted_votes.get(0, 0):.4f}, Classe 1 = {weighted_votes.get(1, 0):.4f}")
    print(f"   ‚Üí D√©cision : Classe {predicted_label} (somme des poids)")
```
